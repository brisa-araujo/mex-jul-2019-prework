```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

import random

game = ["stone","paper","scissors"]

m = 5
w = 3

def computer_play():
    comp_play = random.choice(["stone","paper","scissors"])
    return comp_play

def user_play():
    a = input("Choose 'stone', 'paper' or 'scissors':\n")
    if a == "stone" or a == "paper" or a == "scissors":
        return a
    else:
        print("Choose from the following options: 'stone', 'paper' or 'scissors'.\n")
        a = input("Choose 'stone', 'paper' or 'scissors':\n")

user_choice = 0
computer_choice = 0

def match(a,b):
    a = user_play()
    b = computer_play()
    print("\nUser's choice: %s \nComputer's choice: %s" % (a,b))
    if a == b :
        print("\nIt's a tie")
        return 0
    elif b == "stone" and a == "scissors":
        print ("\nComputer wins")
        return 1
    elif b == "paper" and a == "stone":
        print ("\nComputer wins")
        return 1
    elif b == "scissors" and a == "paper":
        print ("\nComputer wins")
        return 1
    else:
        print("\nUser wins")
        return 2
    

w_comp = 0
w_user = 0
matches = 0

while matches <= m:
    print("\nLet's start a new Rock-Paper-Scissors match.")
    result = match(user_choice, computer_choice)
    if result == 0:
        matches += 1
    elif result == 1:
        w_comp += 1
        matches += 1
    elif result == 2:
        w_user += 1
        matches += 1
    if w_comp == w:
        print("\nComputer wins :)\nComputer: %s \nPlayer: %s" % (w_comp,w_user))
    elif w_user == w:
        print("\nUser wins :)\nPlayer: %s \nComputer: %s" % (w_user,w_comp))
    if w_user == 3 or w_comp == 3:
        break
    elif matches == m:
        print("\nComputer: %s \nPlayer: %s \nNobody won. Good Game!" % (w_user,w_comp))
        break
```
